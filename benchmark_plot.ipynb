```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# CSV file
CSV_FILE = "bk.csv"
OUTPUT_DIR = "plots3"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Read CSV
df = pd.read_csv(CSV_FILE)

# Implementations to plot (skip Python)
implementations = ["grapheasy","graphskew"] #BGL

# Line styles (black & white only)
line_styles = ["-", "--", ":"]

# Matplotlib style tweaks for publication
plt.rcParams.update({
    "font.size": 16,          # base font size
    "axes.labelsize": 18,     # x/y labels
    "axes.titlesize": 20,     # plot titles
    "legend.fontsize": 16,    # legend
    "xtick.labelsize": 14,
    "ytick.labelsize": 14,
    "lines.linewidth": 2.0,   # thicker lines
    "figure.dpi": 300,        # high resolution
})

# Unique vertex sizes
vertex_sizes = sorted(df['Vertices'].unique())

for v in vertex_sizes:
    subset = df[df['Vertices'] == v]

    plt.figure(figsize=(8,6))

    for i, impl in enumerate(implementations):
        impl_data = subset[subset['Implementation'] == impl]
        # Sort by Edges
        impl_data = impl_data.sort_values('Edges')
        plt.plot(
            impl_data['Edges'],
            impl_data['UserTime'],
            line_styles[i % len(line_styles)],
            marker='o',
            markersize=6,
            color="black",
            label=impl
        )

    plt.title(f"BK Runtime vs Edges ({v:,} Vertices)")
    plt.xlabel("Edges")
    plt.ylabel("Time (s)")
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.legend(title="Implementation", loc="best", frameon=True)

    # Save figure
    out_file = os.path.join(OUTPUT_DIR, f"BK_line_{v}_vertices.pdf")
    plt.tight_layout()
    plt.savefig(out_file, dpi=300, bbox_inches="tight")
    plt.close()
    print(f"✅ Saved publication-quality line plot for {v} vertices -> {out_file}")
```
